```python
# import the package and give it the shorter name 'plt'
import matplotlib.pyplot as plt

%matplotlib inline

x = list(range(0,10))
y = list(range(-10, 0))

plt.plot(x,y)
```

```python
# para nomear o gráfico
plt.title("Triangle")
# para nomear o eixo x
plt.xlabel("Array A")
# para nomear o eixo y
plt.ylabel("Array B")
# dados que serao usados para definir o gráfico
plt.axis([-50, 80, 2, 8])
# se quiser alterar os parâmetros
plt.xticks ((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', 'o'))
# para definir a cor
plt.plot(a,b, color='red')
```
