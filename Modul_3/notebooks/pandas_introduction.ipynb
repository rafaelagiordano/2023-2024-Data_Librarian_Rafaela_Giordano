# 2.5 Fallstudie: Feature Engineering

```python
import pandas as pd
df = pd.read_json("../data/Library_Usage.json")
df
```

```python
df['Circulation Active Year']
```

```python
pd.to_numeric(
  df['Circulation Active Year'], errors='coerce'
)
```

```python
df['Circulation Active Month'] = pd.to_datetime(
    df['Circulation Active Month'],
    errors='coerce',
    format="%b"
)
df['Circulation Active Month']
```

```python
df['Circulation Active Month'] = df['Circulation Active Month'].dt.month
df['Circulation Active Month']
```

```python
df
```

```python
df['Year Patron Registered']
```

```python
df['Membership Duration'] = (df['Circulation Active Year'] - df['Year Patron Registered'])
```

```python
df['Membership Duration'] 
```

```python
df['Membership Duration'] = (
    df['Circulation Active Year'] - df['Year Patron Registered']
)*12 + df['Circulation Active Month']
```

```python
df['Membership Duration'] 
```

```python
df.head()
```

```python
df.to_csv('../data/Library_Usage_Membership.csv', index=False)
```

```python
df['Membership Duration'].fillna(0)
```
