```python
cross_tab = pd.crosstab(
    df['materialtype'],
    df['checkouts'],
    margins=True
)
cross_tab
```

```python
cross_tab.columns
```

```python
cross_tab_all = cross_tab['All']
cross_tab_all
```

```python
cross_tab_all.reset_index()
```

```python
type(cross_tab_all)
```

```python
cross_tab_new = cross_tab_all.to_frame()
```

```python
type(cross_tab_new)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(cross_tab_new, annot=True, fmt='d', cmap='Blues', cbar_kws={'label': ' ' })
plt.title('Ausleihen')
plt.xlabel('materialtype')
plt.ylabel('checkouts')
plt.show()
```
