```python
class Node:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

def build_tree():
    # Create nodes for each level
    root = Node("Einzelne Länder")
    
    branch1 = Node("Afrika")
    branch2 = Node("Asien")
    branch3 = Node("Australien")
    branch4 = Node("Europa")
    branch5 = Node("Nordamerika")
    branch6 = Node("Südamerika")

    leaf1A = Node("Marokko")
    leaf2B = Node("Japan")
    leaf3C = Node("Australien")
    leaf4D = Node("Deutschland")
    leaf5E = Node("Kanada")
    leaf6F = Node("Brasilien")
    
    subleaf6F1 = Node("L/Bras: I")
    subleaf6F2 = Node("L/Bras: II")
    subleaf6F3 = Node("L/Bras: III")
    subleaf6F4 = Node("L/Bras: IV")
    subleaf6F5 = Node("L/Bras: V")
    subleaf6F6 = Node("L/Bras: VI")
    subleaf6F7 = Node("L/Bras: VII")
    subleaf6F8 = Node("L/Bras: VIII")
    subleaf6F9 = Node("L/Bras: IX")
    subleaf6F10 = Node("L/Bras: X")
    subleaf6F11 = Node("L/Bras: XI")
    subleaf6F12 = Node("L/Bras: XII")
    subleaf6F13 = Node("L/Bras: XIII")
    subleaf6F14 = Node("L/Bras: XIV")
    subleaf6F15 = Node("L/Bras: XV")
    subleaf6F16 = Node("L/Bras: XVI")
    subleaf6F17 = Node("L/Bras: XVII")
    subleaf6F18 = Node("L/Bras: XVIII")
    subleaf6F19 = Node("L/Bras: XIX")

    # Construct the tree
    root.add_child(branch1)
    root.add_child(branch2)
    root.add_child(branch3)
    root.add_child(branch4)
    root.add_child(branch5)
    root.add_child(branch6)

    branch1.add_child(leaf1A)
    branch2.add_child(leaf2B)
    branch3.add_child(leaf3C)
    branch4.add_child(leaf4D)
    branch5.add_child(leaf5E)
    branch6.add_child(leaf6F)

    leaf6F.add_child(subleaf6F1)
    leaf6F.add_child(subleaf6F2)
    leaf6F.add_child(subleaf6F3)
    leaf6F.add_child(subleaf6F4)
    leaf6F.add_child(subleaf6F5)
    leaf6F.add_child(subleaf6F6)
    leaf6F.add_child(subleaf6F7)
    leaf6F.add_child(subleaf6F8)
    leaf6F.add_child(subleaf6F9)
    leaf6F.add_child(subleaf6F10)
    leaf6F.add_child(subleaf6F11)
    leaf6F.add_child(subleaf6F12)
    leaf6F.add_child(subleaf6F13)
    leaf6F.add_child(subleaf6F14)
    leaf6F.add_child(subleaf6F15)
    leaf6F.add_child(subleaf6F16)
    leaf6F.add_child(subleaf6F17)
    leaf6F.add_child(subleaf6F18)
    leaf6F.add_child(subleaf6F19)
    
    return root

def generate_html(node):
    if not node.children:
        return f'<li>{node.name}</li>'
    children_html = ''.join(generate_html(child) for child in node.children)
    return f'''
    <li>
        <span class="caret">{node.name}</span>
        <ul class="nested">
            {children_html}
        </ul>
    </li>
    '''

tree = build_tree()
html_content = f'''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Collapsible Menu</title>
    <style>
        .nested {{
            display: none;
        }}
        .active {{
            display: block;
        }}
        .caret {{
            cursor: pointer;
            user-select: none;
        }}
    </style>
</head>
<body>
    <h2>Collapsible Tree Menu</h2>
    <ul id="myMenu">
        {generate_html(tree)}
    </ul>

    <script>
        document.addEventListener('DOMContentLoaded', function() {{
            var toggler = document.getElementsByClassName("caret");
            for (var i = 0; i < toggler.length; i++) {{
                toggler[i].addEventListener("click", function() {{
                    this.parentElement.querySelector(".nested").classList.toggle("active");
                    this.classList.toggle("caret-down");
                }});
            }}
        }});
    </script>
</body>
</html>
'''

# Save the HTML content to a file
with open('Test_2.html', 'w') as file:
    file.write(html_content)

print("HTML file 'Test_2.html' has been generated.")
```
