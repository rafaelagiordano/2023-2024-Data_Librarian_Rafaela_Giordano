```python
class Node:
    def __init__(self, title):
        self.title = title
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

    def to_dict(self):
        return {
            'title': self.title,
            'children': [child.to_dict() for child in self.children]
        }
```
