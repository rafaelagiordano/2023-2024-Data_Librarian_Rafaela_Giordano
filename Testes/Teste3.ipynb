```python
class Node:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

def build_tree():
    # Create nodes for each level
    root = Node("Einzelne Länder")
    
    branch1 = Node("Afrika")
    branch2 = Node("Asien")
    branch3 = Node("Australien")
    branch4 = Node("Europa")
    branch5 = Node("Nordamerika")
    branch6 = Node("Südamerika")

    leaf1A = Node("Marokko")
    leaf2B = Node("Japan")
    leaf3C = Node("Australien")
    leaf4D = Node("Deutschland")
    leaf5E = Node("Kanada")
    leaf6A = Node("Lateinamerika")
    leaf6B = Node("Argentinien")
    leaf6C = Node("Bolivien")
    leaf6D = Node("Brasilien")
    leaf6E = Node("Chile")
    leaf6F = Node("Ecuador")
    leaf6G = Node("Guyana")
    leaf6H = Node("Kolumbien")
    leaf6I = Node("Paraguay")
    leaf6J = Node("Peru")
    leaf6K = Node("Suriname")
    leaf6L = Node("Uruguay")
    leaf6M = Node("Venezuela")
    
    subleaf6D1 = Node("L/Bras: I")
    subleaf6D2 = Node("L/Bras: II")
    subleaf6D3 = Node("L/Bras: III")
    subleaf6D4 = Node("L/Bras: IV")
    subleaf6D5 = Node("L/Bras: V")
    subleaf6D6 = Node("L/Bras: VI")
    subleaf6D7 = Node("L/Bras: VII")
    subleaf6D8 = Node("L/Bras: VIII")
    subleaf6D9 = Node("L/Bras: IX")
    subleaf6D10 = Node("L/Bras: X")
    subleaf6D11 = Node("L/Bras: XI")
    subleaf6D12 = Node("L/Bras: XII")
    subleaf6D13 = Node("L/Bras: XIII")
    subleaf6D14 = Node("L/Bras: XIV")
    subleaf6D15 = Node("L/Bras: XV")
    subleaf6D16 = Node("L/Bras: XVI")
    subleaf6D17 = Node("L/Bras: XVII")
    subleaf6D18 = Node("L/Bras: XVIII")
    subleaf6D19 = Node("L/Bras: XIX")

    subsubleaf6D7Aa = Node("L/Bras: VII Aa: Geschichte: Gesamtdarstellung")
    subsubleaf6D7Ab = Node("L/Bras: VII Ab: Geschichte bis 1945")
    subsubleaf6D7Ac = Node("L/Bras: VII Ac: Geschichte ab 1946")

    # Construct the tree
    root.add_child(branch1)
    root.add_child(branch2)
    root.add_child(branch3)
    root.add_child(branch4)
    root.add_child(branch5)
    root.add_child(branch6)

    branch1.add_child(leaf1A)
    branch2.add_child(leaf2B)
    branch3.add_child(leaf3C)
    branch4.add_child(leaf4D)
    branch5.add_child(leaf5E)
    branch6.add_child(leaf6A)
    branch6.add_child(leaf6B)
    branch6.add_child(leaf6C)
    branch6.add_child(leaf6D)
    branch6.add_child(leaf6E)
    branch6.add_child(leaf6F)    
    branch6.add_child(leaf6G)
    branch6.add_child(leaf6H)
    branch6.add_child(leaf6I)
    branch6.add_child(leaf6J)
    branch6.add_child(leaf6K)
    branch6.add_child(leaf6L)
    branch6.add_child(leaf6M)    

    leaf6D.add_child(subleaf6D1)
    leaf6D.add_child(subleaf6D2)
    leaf6D.add_child(subleaf6D3)
    leaf6D.add_child(subleaf6D4)
    leaf6D.add_child(subleaf6D5)
    leaf6D.add_child(subleaf6D6)
    leaf6D.add_child(subleaf6D7)
    leaf6D.add_child(subleaf6D8)
    leaf6D.add_child(subleaf6D9)
    leaf6D.add_child(subleaf6D10)
    leaf6D.add_child(subleaf6D11)
    leaf6D.add_child(subleaf6D12)
    leaf6D.add_child(subleaf6D13)
    leaf6D.add_child(subleaf6D14)
    leaf6D.add_child(subleaf6D15)
    leaf6D.add_child(subleaf6D16)
    leaf6D.add_child(subleaf6D17)
    leaf6D.add_child(subleaf6D18)
    leaf6D.add_child(subleaf6D19)

    subleaf6D7.add_child(subsubleaf6D7Aa)
    subleaf6D7.add_child(subsubleaf6D7Ab)
    subleaf6D7.add_child(subsubleaf6D7Ac)
    
    return root

def generate_html(node):
    if not node.children:
        return f'<li>{node.name}</li>'
    children_html = ''.join(generate_html(child) for child in node.children)
    return f'''
    <li>
        <span class="caret">{node.name}</span>
        <ul class="nested">
            {children_html}
        </ul>
    </li>
    '''

tree = build_tree()
html_content = f'''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Collapsible Menu</title>
    <style>
        .nested {{
            display: none;
        }}
        .active {{
            display: block;
        }}
        .caret {{
            cursor: pointer;
            user-select: none;
        }}
    </style>
</head>
<body>
    <h2>Systematik MPIL</h2>
    <ul id="myMenu">
        {generate_html(tree)}
    </ul>

    <script>
        document.addEventListener('DOMContentLoaded', function() {{
            var toggler = document.getElementsByClassName("caret");
            for (var i = 0; i < toggler.length; i++) {{
                toggler[i].addEventListener("click", function() {{
                    this.parentElement.querySelector(".nested").classList.toggle("active");
                    this.classList.toggle("caret-down");
                }});
            }}
        }});
    </script>
</body>
</html>
'''

# Save the HTML content to a file
with open('collapsible_menu_3.html', 'w') as file:
    file.write(html_content)

print("HTML file 'collapsible_menu_3.html' has been generated.")
```
