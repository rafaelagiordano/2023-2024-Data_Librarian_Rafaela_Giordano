## Trabalhando com DataFrame

```python
#dictionary

people = {
    'first' : ['Corey', 'Jane', 'Anton'],
    'last' : ['Schafer', 'Silva', 'Santos'],
    'email' : ['cs@gmail.com', 'js@gmail.com', 'as@gmail.com']
}
```

```python
df = pd.DataFrame(people)
df
```
