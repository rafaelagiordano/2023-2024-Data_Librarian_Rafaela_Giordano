```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://dl.acm.org/doi/proceedings/10.1145/2814864"
```

```python
request = requests.get(url)
```

```python
soup = BeautifulSoup(request.content, "html.parser")
```

```python
print(soup)
```

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
soup.title.parent.name
```

```python
soup.find_all("a")
```

```python
soup.find_all("h5")
```

```python
soup.find_all("h5", {"class":"issue-item__title"})
```

```python
for publication in soup.find_all("h5", {"class":"issue-item__title"}):
    print(publication.string)
    print( )
```

```python
for publication in soup.find_all("h5", {"class":"issue-item__title"}):
    print(publication.string)
    print(publication.a["href"])
    print( )
```

```python
for publication in soup.find_all("h5", {"class":"issue-item__title"}):
    print(publication.string)
    print(publication.a["href"].replace("doi/", ""))
    print( )
```

```python
for publication in soup.find_all("h5", {"class":"issue-item__title"}):
    print("Title:", publication.string)
    print("DOI:", publication.a["href"].replace("doi/", ""))
    print( )
```
