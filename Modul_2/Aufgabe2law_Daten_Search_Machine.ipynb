```python
import pandas as pd
import requests
```

```python
url_base = "https://www.bibsonomy.org/json/search/"
url_attr = "?items=1000"

def df_query(query):
    url = url_base + query.replace(" ", "%20") + url_attr

    result = requests.get(url)
    json_data = result.json()
    dataframe = pd.DataFrame(json_data["items"])
    return dataframe
```

```python
df = df_query("direito internacional")
```

```python
dataframes = []

queries = ["constitutional law", "administrative law", "human rights law"]

for q in queries:
    df = df_query(q)
    dataframes.append(df)

    print(f'Ergebnisse für {q} gefunden. Dataframe enthält {len(df["id"])} Einträge.')
```

```python
data = pd.concat(dataframes, ignore_index=True)
```

```python
import pyterrier as pt
import pickle
import nltk
import numpy as np
```

```python
if not pt.started():
    pt.init()
```

```python
select['docno'] = select['id']
select['text'] = select['label']
```

```python
index_ordner = './law_index'

indexer = pt.DFIndexer(index_ordner, overwrite=True)
index_ref = indexer.index(select['text'], select['docno'])

print(f"path to our index: {index_ref.toString()}")
```

```python
index = pt.IndexFactory.of(index_ref)
```

```python
def get_idf_for_term(term,index):
    lex = index.getLexicon()
    stemmed_term = stemmer.stem(term)
    if not stemmed_term in lex:
        return
   
    lex_entry = lex[stemmed_term]
   
    df_term = lex_entry.getDocumentFrequency()
    N = index.getCollectionStatistics().numberOfDocuments
    
    idf = N/df_term
   
    log_idf = np.emath.logn(10, idf)
    
    return log_idf
```
