```python
# csv-Datei als Text behandeln
for line in open("openapc.csv", "r"):
    row = line.split(",") #Zeile a, Trennzeichen splitten
    institution = row[0] #Zuordnung der Felder
    period = row[1]
    journal_full_title = row[6]
    issn = row[7].strip()
    print("institution: {}\nperiod: {}\njournal_full_title: {}\nissn: {}\n".format(institution, period, journal_full_title, issn))
```

```python
import csv
```

```python
data = []
with open(path, "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    for row in reader:
        data.append(row)
```

```python
# einlesen mit DicReader
with open(path, "r") as csvfile:
    reader = csv.DictReader(csvfile, delimiter=",")
    for row in reader:
        print("Title: {}\nAuthor: {}\nYear: {}\n".format(row["Title"], row["Author"], row["Year"]))
```

```python
header=["institution", "period", "journal_full_title", "issn"]

with open("openapc.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile, delimiter=",", fieldnames=header)
    for row in reader:
        print("institution: {}\nperiod: {}\njournal_full_title: {}\nissn: {}\n".format(institution, period, journal_full_title, issn))
```
