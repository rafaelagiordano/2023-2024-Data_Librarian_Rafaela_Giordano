```python
import pandas as pd
import requests
url = "https://www.bibsonomy.org/json/search/portuguese?items=1000"
result = requests.get(url)
data = result.json()
```

```python
data.keys()
```

```python
type(data["items"])
```

```python
df_bibsonomy = pd.DataFrame(data["items"])
```

```python
df_bibsonomy[['type', 'label', 'date', 'url', 'journal', 'year', 'author', 'abstract', 'language', 'issn', 'publisher']]
```

```python
df_bibsonomy_selection = df_bibsonomy[['type', 'label', 'date', 'url', 'journal', 'year', 'author', 'publisher', 'isbn']]
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.loc[df_bibsonomy_selection["author"].isnull() == False]
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.loc[df_bibsonomy_selection["publisher"].isnull() == False]
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection.loc[df_bibsonomy_selection.type == "Publication"]
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.drop_duplicates(subset = "label", keep = "first")
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection.loc[df_bibsonomy_selection["year"] > 2005]
```

```python
df_bibsonomy_selection["year"].unique()
```

```python
df_bibsonomy_selection.loc[df_bibsonomy_selection.year.astype(float) >= 2005]
```

```python
df_bibsonomy_selection["year_number"] = df_bibsonomy_selection["year"].astype(int)
```

```python
df_bibsonomy_selection["year_correct"] = df_bibsonomy_selection.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection.loc[df_bibsonomy_selection.year_correct == False]
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.loc[df_bibsonomy_selection.year_correct == True]
```

```python
df_bibsonomy_selection.loc[df_bibsonomy_selection.year.astype(int) >=2000]
```

```python
df_bibsonomy_selection["year_number"] = df_bibsonomy_selection["year"].astype(int)
```

```python
df_bibsonomy_selection.info()
```

```python
df_bibsonomy_selection.loc[df_bibsonomy_selection.year_number >=2000]
```

```python
df_bibsonomy_selection_recent = df_bibsonomy_selection.loc[df_bibsonomy_selection.year_number >=2000]
```

```python
df_bibsonomy_selection_recent
```

```python
df_bibsonomy_selection_final = df_bibsonomy_selection_recent[['type', 'label', 'date', 'url', 'year', 'author', 'publisher']]
```

```python
df_bibsonomy_selection_final
```

```python
df_bibsonomy_selection_final = df_bibsonomy_selection_final.rename(columns={"label": "title", "author": "authors"})
```

```python
df_bibsonomy_selection_final
```

```python
import json
df_bibsonomy_selection_final.to_json("output_bibsonomy.json", orient="records")
```
